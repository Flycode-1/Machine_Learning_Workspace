### Applying HAAR cascading for Eye detection

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
#Loading image to detect
img = cv2.imread("faces.jpeg",1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

disp_img = img[:,:,::-1]

plt.axis("off")
plt.imshow(disp_img)
plt.tight_layout()
plt.show()
```

```python
#Loading our model 
path = "haarcascade_eye.xml"

eye_cascade = cv2.CascadeClassifier(path)

#applying our model
eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.02,minNeighbors=20,minSize=(10,10))

#Draw circle for eyes detected
for (x, y, w, h) in eyes:
    
    xc = (x + x+w)/2
    yc = (y + y+h)/2
    radius = w/2
    cv2.circle(img, (int(xc),int(yc)), int(radius), (255,0,0), 2)
    
plt.axis("off")
plt.imshow(img[:,:,::-1])
plt.figure(figsize=(1,1))
plt.tight_layout()
plt.show()
```

Ajusting the parameter and switching for a better trained model can improve the results , but the result is pretty good given the basic model we've used ! '

## Applying HAAR for face detection

```python
img = cv2.imread("faces.jpeg",1)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
path = "haarcascade_frontalface_default.xml"

face_cascade = cv2.CascadeClassifier(path)

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.10, minNeighbors=5, minSize=(40,40))
print(len(faces))

for (x, y, w, h) in faces:
	cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
plt.imshow(img)
```
